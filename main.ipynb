```python
import numpy as np

# Assuming cols4 is a list of column names
# Add your column names here
cols3 = ['Year','Access to electricity (% of population)','Access to clean fuels for cooking','Renewable-electricity-generating-capacity-per-capita','Renewable energy share in the total final energy consumption (%)','Electricity from fossil fuels (TWh)','Electricity from nuclear (TWh)','Electricity from renewables (TWh)','Primary energy consumption per capita (kWh/person)','Energy intensity level of primary energy (MJ/$2017 PPP GDP)','Value_co2_emissions_kt_by_country','gdp_growth','gdp_per_capita','Land Area(Km2)','Latitude','Longitude']
# Create an array to store means
mean = np.zeros(10)  # Initialize with zeros, assuming you want an array of length 10

# Calculate means for specific columns in df_mean
for k in range(1, min(10, len(cols3))):
    mean[k] = df_mean[cols3[k]].mean()

# Fill missing values in new columns based on the means
for k in range(1, min(10, len(cols4))):
    column_name = f"mean_{cols3[k]}"
    mean_fill = mean[k]  # Mean for this specific column

    df_mean[column_name] = df_mean[cols3[k]].fillna(mean_fill)
```

```python
import numpy as np

# Assuming cols4 is a list of column names
# Add your column names here
cols3 = ['Year','Access to electricity (% of population)','Access to clean fuels for cooking','Renewable-electricity-generating-capacity-per-capita','Renewable energy share in the total final energy consumption (%)','Electricity from fossil fuels (TWh)','Electricity from nuclear (TWh)','Electricity from renewables (TWh)','Primary energy consumption per capita (kWh/person)','Energy intensity level of primary energy (MJ/$2017 PPP GDP)','Value_co2_emissions_kt_by_country','gdp_growth','gdp_per_capita','Land Area(Km2)','Latitude','Longitude']
# Create an array to store means
median = np.zeros(10)  # Initialize with zeros, assuming you want an array of length 10

# Calculate means for specific columns in df_mean
for k in range(1, min(10, len(cols3))):
    median[k] = df_median[cols3[k]].median()

# Fill missing values in new columns based on the means
for k in range(1, min(10, len(cols3))):
    column_name = f"median_{cols3[k]}"
    median_fill = median[k]  # Mean for this specific column

    df_median[column_name] = df_median[cols3[k]].fillna(median_fill)
```
